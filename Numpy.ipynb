# Statistic

```python
# Standard variation = +- square root of variance
# variance = sum square difference from Mean, 
# what is normal, what is large, extra large, extra small from normal - MEAN of collection
```

```python
dog_height=[600,470,170,430,300]
```

```python
dog_height=np.array(dog_height)
np.std(dog_height)
```

```python
np.var(dog_height)
```

```python
np.sqrt(np.var(dog_height))
```

```python
np.random.seed(0)
sales_amount=np.random.randint(20,size=(5,3))
sales_amount
```

```python
import pandas as pd
weekly_sales=pd.DataFrame(sales_amount,index=["Mon","Tue","Wed","Thu","Fri"],
                         columns=["Almond Butter","Peanut Butter","Cashew Butter"])
```

```python
weekly_sales
```

```python
# create price array
prices=np.array([10,8,12])
butter_prices=pd.DataFrame(prices.reshape(1,3),index=["Prices"],
                         columns=["Almond Butter","Peanut Butter","Cashew Butter"])

```

```python
butter_prices
```

```python
total_sales=weekly_sales.dot(butter_prices.T)
```

```python
total_sales
```

```python
weekly_sales["Total"]=total_sales
weekly_sales
```

```python
rng=np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)
sizes=1000*rng.rand(100)
colors=rng.rand(100)
fig,ax=plt.subplots()
img1=ax.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.3)
fig.colorbar(img1)
```

```python
SOFT_DRINK_PRICES={"COOKE":10,
                   "pepsi":12,
                   "dr thanh":8}
fig,ax=plt.subplots();
ax.bar(SOFT_DRINK_PRICES.keys(),SOFT_DRINK_PRICES.values())
ax.set(title="BHX Gia do uong",ylabel="Prices")
```

```python
fig,ax=plt.subplots();
ax.barh(list(SOFT_DRINK_PRICES.keys()),list(SOFT_DRINK_PRICES.values()))
ax.set(title="BHX Gia do uong",xlabel="Prices")
```

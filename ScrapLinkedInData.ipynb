```python
def GetURLs():
    page_source=BeautifulSoup(driver.page_source)

    #print(page_source)
    profiles=page_source.find_all('a',class_="app-aware-link")
    #print(profiles)
    all_profiles_URL=[]
    for profile in profiles:
        profile_URL=profile.get('href')
        if 'miniProfileUrn' in profile_URL:
            all_profiles_URL.append(profile_URL)
    set_all_profiles_URL=set(all_profiles_URL)
    #print(set_all_profiles_URL)
    #print(len(set_all_profisles_URL))
    return set_all_profiles_URL

```

```python
print(GetURLs())
```

TASK 4: MOVING TO NEXT RESULT PAGE

```python
from selenium.webdriver.common.by import By
```

```python
def GetURLsonPage():
    numberofpage=input("How many pages that you want to search and collect profiles?")
    num=int(numberofpage)
    all_profiles=[]
    for page in range(num):
        URL_one_page=GetURLs()
        #if URL_one_page not in all_profiles:
        all_profiles+=URL_one_page
        sleep(2)
        driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
        sleep(2)
        next_button=driver.find_element(By.CLASS_NAME,'artdeco-pagination__button--next')
        driver.execute_script("arguments[0].click();", next_button)
        sleep(3)
    return set(all_profiles)

#print(GetURLsonPage())

URL_all_pages=GetURLsonPage()


        
```

WRITE OUTPUT TO CSV FILE

```python
with open ('output.csv','w',encoding='UTF8', newline='') as file_output:
    headers=['Name','Job Title','Company', 'Location', 'PROFILE URL']
    writer=csv.writer(file_output)
    writer.writerow(headers)

    for linkedinURL in URL_all_pages:
        driver.get(linkedinURL)
        sleep(1)
        page_source=BeautifulSoup(driver.page_source,'html.parser')
        sleep(1.5)
        info_div=page_source.find('div',class_='ph5 pb5')
        #print(info_div)
        info_name=info_div.find('h1',class_='text-heading-xlarge inline t-24 v-align-middle break-words')
        info_position=info_div.find('div',class_='text-body-medium break-words')
        info_company=info_div.find('div',class_='inline-show-more-text inline-show-more-text--is-collapsed inline-show-more-text--is-collapsed-with-line-clamp inline')
        info_location=info_div.find('span',class_='text-body-small inline t-black--light break-words')
        #print(info_loc)
        if info_name is not None:
            name=info_name.get_text().strip()
        else:
            name='NA'
        if info_position is not None:
            position=info_position.get_text().strip()
        else:
            position='NA'
        if info_company is not None:
            company=info_company.get_text().strip()
        else:
            company='NA'
        if info_location is not None:
            location=info_location.get_text().strip()
        else:
            location='NA'
        
        print(name,position,company,location,linkedinURL)
        data=[name,position,company,location,linkedinURL]
        # write data to csv 1 row
        writer.writerow(data)

```
